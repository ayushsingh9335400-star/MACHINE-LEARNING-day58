```python
import scipy.cluster.hierarchy as shc  
import matplotlib.pyplot as plt

# Example data (replace with your dataset)
import numpy as np
data = np.random.rand(10, 2)  # 10 samples, 2 features
```

```python
# Plot dendrogram
plt.figure(figsize=(10, 7))  # ✅ 'figuresize' → 'figsize'
plt.title("Customer Dendrograms")  # ✅ optional: fixed capitalization
dend = shc.dendrogram(shc.linkage(data, method='ward'))

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Create Agglomerative Clustering model
cluster = AgglomerativeClustering(
    n_clusters=5,
    metric='euclidean',   # ✅ use 'metric' instead of 'affinity'
    linkage='ward'
)

# Fit model and get cluster labels
labels = cluster.fit_predict(data)

print("Cluster labels:", labels)
```

```python
labels
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))  
plt.scatter(data[:, 0], data[:, 1], c=cluster.labels_, cmap='rainbow', s=50)
plt.title("Agglomerative Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
